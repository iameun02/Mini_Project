# I. Data Preprocessing

```
import warnings
warnings.filterwarnings('ignore')
```

> ## 1) 'notExercise.xlsx' 파일 읽어오기

```
import pandas as pd

url = 'https://raw.githubusercontent.com/rusita-ai/pyData/master/notExercise.xlsx'
DF = pd.read_excel(url)

DF.head(10)
```

* DataFrame 정보

```
DF.info()
```

> ## 2) '기간' 열(Column) 삭제

* '기간'열(Column) 정보

```
DF['기간'].info()
```

* '기간'열(Column) <font color = 'red'>삭제

```
DF = DF.drop(columns = ['기간'])
```

* <font color = 'red'>삭제</font> 결과 확인

```
DF[:5]
```

> ## 3) 0, 23~52번(지역분류) index 삭제

* '대분류' 정보

```
DF['대분류'].value_counts()
```

* 23번 ~ 52번 index <font color = 'red'>삭제

```
DF.drop(DF.index[23:53], inplace=True)
```

```
DF
```

* 0번 index <font color = 'red'>삭제

```
DF = DF.iloc[1:,:]
```

```
DF
```

# II. '성별' 통계

> ## 1) '성별' 데이터 추출

```
DF_g =DF[DF['대분류'] == '성별']
```

```
DF_g[:2]
```

> ## 2) '대분류' 열(Column) 삭제

```
DF_g = DF_g.drop(columns=['대분류'])
```

```
DF_g[:2]
```

> ## 3) '분류'열(Column)을 index로 지정

```
DF_g = DF_g.set_index('분류')
```

```
DF_g
```

> ## 4) matplotlib 시각화

```
DF_g.index
```

```
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (7, 7))

ax[0, 0].set_title('운동을 할 충분한 시간이 없어서')
ax[0, 0].pie(DF_g['운동을 할 충분한 시간이 없어서'],
         labels = DF_g.index,
         autopct = '%.1f%%',
         explode = [0.0, 0.0],
         startangle = 90)


ax[0, 1].set_title('함께 운동을 할 사람이 없어서')
ax[0, 1].pie(DF_g['함께 운동을 할 사람이 없어서'],
         labels = DF_g.index,
         autopct = '%.1f%%',
         explode = [0.0, 0.0],
         startangle = 90)

ax[1, 0].set_title('운동을 할 만한 장소가 없어서')
ax[1, 0].pie(DF_g['운동을 할 만한 장소가 없어서'],
         labels = DF_g.index,
         autopct = '%.1f%%',
         explode = [0.0, 0.0],
         startangle = 90)

ax[1, 1].set_title('운동을 싫어해서')
ax[1, 1].pie(DF_g['운동을 싫어해서'],
         labels = DF_g.index,
         autopct = '%.1f%%',
         explode = [0.0, 0.0],
         startangle = 90)

plt.show()

```
